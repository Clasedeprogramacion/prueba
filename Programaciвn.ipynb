# Loops (iteraciones) while

La tercera y última estructura lógica es la iteración, en ella se controla cuantas veces en sucesión una operación o secuencia de operaciones es realizada. Utilizando una estructura de iteración no se necesita definir por ejemplo una declaración *print* 100 veces. Simplemente se le pide al computador ejecutar la declaración *print* cualquier número de veces.

Una de estas estructuras de iteración es la estructura *while*, la cual tiene la sintaxis:

```python
while condicion-para-iterar:
    Operaciones
    A
    Ejeutar
```

```python
# Realizar 100 print! e infinitos
count = 1
while count < 101:
    print "algo", count
    #count = count + 1
    count += 1
```

```python
# Sumatoria
sum = 0
n = 0
while n < 10:
    sum += n
    n += 1
print sum
```

```python
# Factorial n!
factorial = 1
n = 7
while n >= 1:
    factorial *= n
    n -= 1
print factorial
```

## Un loop también se puede controlar con un valor llamado *sentinela*, en este caso para un programa que suma números:

```python
usuarioStr = "Entre un entero (El programa termina si se introduce 0): "
data = input(usuarioStr)

sum = 0
while data != 0:
    sum += data
    data = input(usuarioStr)
print "La suma es", sum
```
